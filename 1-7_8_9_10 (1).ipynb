# 8: IF ELSE AND ELIF (SELECTION OR DECISION CONROL STATEMENTS )

    Normal control flow is Sequencial
    another form is Repetition in case of for and while loops

```python
#IF

rate=35000
if rate<30000:
    print("I can buy that mobile")
```

```python
#IF ELSE

if rate<30000:
    print("I can buy that mobile")
else:
    print("I can't buy it")
```

```python
# ELIF

amount=int(input("Enter Your Amount: "))

if amount<1000:
    discount = amount*0.05
    print("Discount: ",discount)
elif amount<5000:
    discount = amount*0.10
    print("Discount",discount)
else:
    discount = amount*0.15
    print("Discount",discount)

print("Net Payable: ",amount-discount)
```

```python
# NESTED IF

num=0
if num<6:
    if num==1:
        print("First Prize")
    elif num==2:
        print("Second Prize")
    elif num==3:
        print("Third Prize")
    elif num==4:
        print("Fourth Prize")
    elif num==5:
        print("Fifth Prize")
    else:
        print("No Prizes")
```

```python
#USAGE OF BREAK STATEMENT

i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1      # +=,-=,*= etc usage
```

```python
#USAGE OF CONTINUE STATEMENT

i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
# WHILE ELSE

i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
